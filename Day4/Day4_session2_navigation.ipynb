### Exercies

Print each directory in the MPI-workshop folder, and list their content beneath them with a tab

```python
for dirname in os.listdir():
    print(dirname)
    try:
        os.chdir(dirname)
        for filename in glob.glob("*.ipynb"):
            print('\t' + filename)
        os.chdir('..')
        print('----------------------------')
    except NotADirectoryError:
        print('Hello')
```
