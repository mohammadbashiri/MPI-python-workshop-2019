### Anatomy of matplotlib plot

In essence, there are two big components that you need to take into account:

- The **Figure** is the overall window or page that everything is drawn on. It’s the top-level component of all the ones that you will consider in the following points. You can create multiple independent Figures. A Figure can have several other things in it, such as a **suptitle**, which is a centered title to the figure. You’ll also find that you can add a **legend** and **color bar**, for example, to your Figure.

- To the figure you add Axes. The Axes is the area on which the data is plotted with functions such as `plot()` and `scatter()` and that can have ticks, labels, etc. associated with it. This explains why Figures can contain multiple Axes.

```python
x = np.deg2rad(np.arange(361))
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin, lw=10, ls='--', color=(55/255, 199/255, 166/255), label='Sin')
plt.plot(x, y_cos, lw=10, ls=':', color=(1, 132/255, 24/255), label='Cos')

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
# ax.spines['bottom'].set_visible(False)
# ax.spines['left'].set_visible(False)

plt.ylim(-2, 2)

plt.xlabel("X label", fontsize=30, color='r')
plt.ylabel('Y label', fontsize=30)

plt.xticks([1, 2.5, 5], "Python is Awesom!".split(' '), fontsize=20, fontfamily='serif', rotation=-45)
plt.yticks(fontsize=20)
plt.locator_params(axis='y', nbins=5)


plt.legend(loc="lower left", fontsize=15, facecolor='.8')
plt.title("Title", y=1.05, fontsize=50)
plt.grid(linestyle='dashed')
```

```python
fig, ax = plt.subplots(figsize=(10, 8), facecolor='r', edgecolor='k', linewidth=10)

ax.set_facecolor((0, 0, 1))

# remove borders
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
# ax.spines['bottom'].set_visible(True)
# ax.spines['left'].set_visible(True)

# data
x = np.deg2rad(np.arange(361))
y_sin = np.sin(x)
y_cos = np.cos(x)

ax.plot(x, y_sin, lw=10, ls='dashed', color=(85/255, 212/255, 195/255), label='Sin')
ax.plot(x, y_cos, lw=10, ls=':', color=(247/255, 8/255, 120/255), label='Cos')

ax.set_ylim(-2, 2)

ax.set_xlabel("X label", fontsize=30)
ax.set_ylabel("y label", fontsize=30, rotation=0)

ax.xaxis.set_label_coords(1.2, .05)
ax.yaxis.set_label_coords(-.25, .42)

ax.set_xticks([1, 2.5, 5])
ax.set_xticklabels("Python is Awesome!".split(' '), color='r', fontsize=30, rotation=-45)
ax.tick_params(axis='y', labelsize=20)

ax.locator_params(axis='y', nbins=5)

ax.legend(loc="lower left", fontsize=15)
ax.set_title("Title", y=1.05, fontsize=50)
ax.grid()

fig.savefig('our_figure.jpg', bbox_inches='tight', facecolor='r')
```

Pretty much whatever we did above can be done with the axis objec as well. In other words, the methods used to manipulate the properties of a plot, for example `plt.xlabel()` all exist for an axes object as well. For instacne, instead of `plt.xlabel()` we can also use `ax.set_xlabel()`.  

```python
plt.figure()
```

```python
fig, ax = plt.subplots(figsize=(10, 7), facecolor='.9', edgecolor='k', linewidth=10)
ax.set_facecolor((250/255, 230/255, 110/255))

# remove borders
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
# ax.spines['bottom'].set_visible(True)
# ax.spines['left'].set_visible(True)

# data
x = np.deg2rad(np.arange(361))
y_sin = np.sin(x)
y_cos = np.cos(x)

ax.plot(x, y_sin, lw=10, ls='dashed', color=(85/255, 212/255, 195/255), label='Sin')
ax.plot(x, y_cos, lw=10, ls=':', color=(247/255, 8/255, 120/255), label='Cos')

# changing the axes
ax.set_ylim(-2, 2)

# add x and y labels
ax.set_xlabel('X label', fontsize=30)
ax.set_ylabel('Y label', position=(1, .5), fontsize=30, color='.7', rotation=0)

# change the location of labels
ax.xaxis.set_label_coords(1.1, 0.03)
ax.yaxis.set_label_coords(-.15, .5)

# change ticks and ticklabels
ax.set_xticks([1, 2.7, 5])
ax.set_xticklabels(['Python', 'Is', 'Awesome'], color=(247/255, 8/255, 120/255), rotation=-45)
ax.tick_params(axis='both', direction='in', labelsize=30)

# change the number of bins for y-axis
ax.locator_params(axis='y', nbins=5)

ax.set_title("Title", fontsize=50, y=1.05)

ax.grid()
ax.legend(loc="lower left", fontsize=20)

plt.show()
```

### Using gridpace

```python
from matplotlib.gridspec import GridSpec

fig = plt.figure(figsize=(15, 10))

gs = GridSpec(3, 3)
ax1 = plt.subplot(gs[0, :])
ax2 = plt.subplot(gs[1, :-1])
ax3 = plt.subplot(gs[1:, -1])
ax4 = plt.subplot(gs[2, 0])
ax5 = plt.subplot(gs[2, 1])

axes = (ax1, ax2, ax3, ax4, ax5)

# and then the same story
```

Ultimate control: Create axes separately, set their size and put them wherever you want..

```python
def pl(ax):
    ax.set_facecolor(*np.random.uniform(size=(1, 3)))
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    
    # data
    x = np.deg2rad(np.arange(361))
    y_sin = np.sin(x)
    y_cos = np.cos(x)

    ax.plot(x, y_sin, ls='dashed', color=(85/255, 212/255, 195/255), label='Sin')
    ax.plot(x, y_cos, ls=':', color=(247/255, 8/255, 120/255), label='Cos')
    
    return ax
```

```python
fig = plt.figure(figsize=(7, 7)) #, dpi=300)

ax1 = fig.add_axes([0.1, 0.1, 0.5, 0.5])
ax2 = fig.add_axes([0.7, 0.1, 0.3, 0.3])
ax3 = fig.add_axes([0.5, 0.4, 0.15, 0.15])


ax1 = pl(ax1)
ax2 = pl(ax2)
ax3 = pl(ax3)
```
